```python
df1 = pd.read_csv('../database/ibge-mas-10000.csv')
df2 = pd.read_csv('../database/ibge-fem-10000.csv')
df3 = pd.concat([df1, df2], ignore_index=True)
df3
```

```python
import random
from datetime import datetime, timedelta

sobrenomes = ["Silva","Oliveira","Souza","Santos","Lima","Pereira","Costa","Ferreira","Rodrigues","Almeida","Nascimento","Ribeiro","Cunha","Martins","Lopes","Mendes","Barbosa","Pinto","Freitas","Araújo","Fernandes","Carvalho","Gomes","Moreira", "Cardoso","Monteiro","Teixeira","Barros","Moura","Dias"]

start_date = datetime(1935, 1, 1)
end_date = datetime(2010, 1, 1)

def random_date(start_date, end_date):
    return start_date + timedelta(
        seconds=random.randint(0, int((end_date - start_date).total_seconds()))
    )

def sobrenomes_aleatorios(linha):
    sobrenome = random.choice(sobrenomes)
    nascimento = random_date(start_date, end_date).strftime("%Y-%m-%d")
    peso = round(random.uniform(50, 120), 2)
    pressao_arterial = f"{random.randint(80, 200)}/{random.randint(30, 150)}"
    glicose = random.randint(45, 250)

    return sobrenome,nascimento, peso, pressao_arterial, glicose

df3["ID"] = range(1, 20001)

def idade_frequencia(linha):
    idade = (datetime.now() - pd.to_datetime(linha['NASCIMENTO'])).days // 365
    frequencia_cardiaca = random.randint(50, 110) - (idade // 6)
    frequencia_cardiaca = max(frequencia_cardiaca, 40)
    altura = round(random.uniform(1.5, 2.2), 2) 

    return idade, frequencia_cardiaca, altura


df3[['SOBRENOME', 'NASCIMENTO', 'PESO', 'PRESSAO_ARTERIAL', 'GLICEMIA']] = df3.apply(sobrenomes_aleatorios, axis=1, result_type='expand')

df3[['IDADE', 'FREQUENCIA_CARDIACA', 'ALTURA']] = df3.apply(idade_frequencia, axis=1, result_type='expand')
df3
```

```python
df4 = df3[['ID','nome', 'SOBRENOME', 'IDADE', 'NASCIMENTO', 'ALTURA', 'PESO','sexo', 'PRESSAO_ARTERIAL', 'GLICEMIA', 'FREQUENCIA_CARDIACA']]
df4
```

```python
df4.columns = df4.columns.str.upper()
df4
```
